# Afficher matrice de correlation

```python
import pandas as pd
```

```python
pd.set_option('display.max_rows', 4)    # Afficher toutes les lignes
pd.set_option('display.max_columns', None) # Afficher toutes les colonnes
pd.set_option('display.width', None)       # Ajuster la largeur de l'affichage
```

```python
labels = []

for file in ['msd_beatunes_map.cls']:
    
    file_path = f'./{file}'

    with open(file_path, 'r') as cls_file:
        for _ in range(7):  # Ignorer les 7 premières lignes
            next(cls_file)

        for line in cls_file:
            parts = line.strip().split()
            track_id = parts[0]
            genres = parts[1:]  # Prendre tous les éléments après le premier comme genres

            labels.append({'track_id': track_id, 'genre': genres[0]})

labels_df = pd.DataFrame(labels)
```

```python
df = pd.read_csv('h5_datas.csv').drop(columns=['genre'])

merged_df = pd.merge(df, labels_df, how='left', on='track_id')
```

```python
# Display track with genre !
genres = list(merged_df['genre'].unique())[1:]

merged_df.loc[merged_df['genre'].isin(genres)]
```

```python
temp_merged_df = merged_df.select_dtypes(include=['int', 'float'])
temp_merged_df.corr()
```
