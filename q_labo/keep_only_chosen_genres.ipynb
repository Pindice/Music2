```python
import pandas as pd

df = pd.read_csv('filtered_with_label.csv')
```

```python
chosen_genres = ['Rock', 'Pop', 'Hip-Hop', 'Country', \
                 'Jazz', 'Latin', 'Reggae', 'R&B', 'Electronic']

df = df.loc[df['genre'].isin(chosen_genres),:]
```

```python
df['genre'].unique()
```

```python
df.mask(df['genre'].isin)
```

```python
count_df = df['genre'].value_counts()
count_df
```

```python
min_df = count_df.min()
balanced_data = df.groupby('genre').apply(lambda x: x.sample(min_df))
balanced_data = balanced_data.reset_index(drop=True)

balanced_data['genre'].value_counts()
```

# Label encode

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

balanced_data['genre'] = label_encoder.fit_transform(balanced_data['genre'])
```

# Save

```python
balanced_data.to_csv('balanced_data_label_encoded.csv', index=False)
```
