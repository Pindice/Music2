```python
from sklearn.feature_selection import chi2
import pandas as pd

df = pd.read_csv('../csv/4_balanced_data_label_encoded.csv')
```

```python
import matplotlib.pyplot as plt

i=1
plt.figure(figsize=(15,5))
for feature in ['key_confidence', 'time_signature_confidence', 'mode_confidence']:
    plt.subplot(1,3,i)
    plt.boxplot(df[feature])
    plt.title(feature + " boxplot")
    i+=1
plt.show()
```

```python
# Considérant les valeurs liée au "confidences" non pertinentes car dont la précision est autour de 50%

df = df.drop(columns=['key', 'key_confidence', 'mode', 'mode_confidence', 'time_signature', 'time_signature_confidence'])

df.to_csv('../csv/5_balanced_data_label_encoded.csv', index=False)
```
