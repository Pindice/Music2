```python
import matplotlib.pyplot as plt
import pandas as pd

pd.set_option('display.max_rows', 13)    # Afficher toutes les lignes
pd.set_option('display.max_columns', None) # Afficher toutes les colonnes
pd.set_option('display.width', None)       # Ajuster la largeur de l'affichage
```

```python
df = pd.read_csv('../csv/balanced_data_label_encoded.csv').drop(columns=['track_id'])
df
```

## Variance

```python
df.var()
```

```python
linear_features = ['duration','end_of_fade_in', 'key_confidence', 'loudness', 'mode_confidence', 'start_of_fade_out', 'tempo', \
            'time_signature_confidence']

categorical_features = ['key', 'mode', 'time_signature']

for feature in linear_features:

    print(feature, " variance =>", df[feature].var(axis=0))
    plt.figure(figsize=(20,5))
    plt.title(feature + " variance")
    plt.plot(df[feature])
    plt.show()

for feature in categorical_features:

    print(feature, " variance =>", df[feature].var(axis=0))
    plt.figure(figsize=(20,5))
    plt.title(feature + " variance")
    plt.hist(df[feature])
    plt.show()
```
