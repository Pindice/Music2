```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('./1_label_encoded.csv')
# df.drop(columns=['label'], inplace=True)
df
```

```python
print(df.columns.values)
```

```python
# Normalize
standard_scaler = StandardScaler()

columns = ['zcr', 'spectral_c', 'rolloff', 'mfcc1', 'mfcc2', 'mfcc3', 'mfcc4', 'mfcc5',
 'mfcc6', 'mfcc7', 'mfcc8', 'mfcc9', 'mfcc10', 'mfcc11', 'mfcc12', 'mfcc13',
 'mfcc14', 'mfcc15', 'mfcc16', 'mfcc17', 'mfcc18', 'mfcc19', 'mfcc20']

for column in columns:
    current_column = df[column].to_numpy()
    current_column = current_column.reshape(1000,1)

    current_column = standard_scaler.fit_transform(current_column)
    df[column] = current_column
    # print(current_column)

df
```

```python
df.to_csv('./2_normalized_with_standard_scaler', index=False)
```
