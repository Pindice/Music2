```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("./features_3_sec.csv").drop(columns=['filename', 'length'])
df
```

```python
list(df.columns)[0:-1]
```

```python
encoder = LabelEncoder()
standard_scaler = StandardScaler()

# Encode labels
encoded_y_numpy = encoder.fit_transform(df['label'].to_numpy())

# Noramlize data
for column in list(df.columns)[0:-1]:
    current_column = df[column].to_numpy()
    current_column = current_column.reshape(9990,1)

    current_column = standard_scaler.fit_transform(current_column)
    df[column] = current_column

# Concat with the original df
concat_df = pd.concat([df, pd.DataFrame(encoded_y_numpy)], axis=1)

# Remove old label column
final_df: pd.DataFrame = concat_df.drop(columns=['label']).rename(columns={0: 'label'})
final_df
```

```python
final_df_bis = final_df.drop(columns=['perceptr_mean', 'perceptr_var'])
final_df_bis
```
